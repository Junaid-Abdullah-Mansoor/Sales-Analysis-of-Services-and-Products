```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("SouthlakeService.csv")
df=df.dropna(axis='columns', how='all')
df=df[2:]
df=df.reset_index()
del df['index']
```

```python
df2=pd.read_csv("SouthlakeProduct.csv")
df2=df2.dropna(axis='columns', how='all')
```

```python
df1=pd.DataFrame(columns = ['Main-Category', 'Sub-Category', 'Description','QTY','Amount','% Cat Sales','% Total Sales'])
for i in range (0,len(df)):
    if (not(pd.isna(df["Services Sales"][i]))):
        if (((df["Services Sales"][i]).find(':')!=-1) and ((df["Services Sales"][i-1]).find(':')==-1)):
            x=df["Services Sales"][i-1]
        if ((df["Services Sales"][i]).find(':')!=-1):
            y=df["Services Sales"][i].split(':')[0]
            z=df["Services Sales"][i].split(':')[1]
            df1=df1.append({'Main-Category':x, 'Sub-Category':y, 'Description':z,'QTY':df['Qty'][i],'Amount':df['Amount'][i],'% Cat Sales':df['% Cat\nSales'][i],'% Total Sales':df['% Total\nSales'][i]},ignore_index=True)
    

```

```python
df3=pd.DataFrame(columns = ['Product-Category', 'Product-Sub-Category', 'Description','QTY','Amount','% Cat Sales','% Total Sales'])
for i in range (1,len(df2)):
    if (not(pd.isna(df2["Products - Southlake"][i])) and (not(pd.isna(df2["Products - Southlake"][i-1])))):
        if (((df2["Products - Southlake"][i]).find(':')!=-1) and ((df2["Products - Southlake"][i-1]).find(':')==-1)):
            x=df2["Products - Southlake"][i-1]
        if ((df2["Products - Southlake"][i]).find(':')!=-1):
            y=df2["Products - Southlake"][i].split(':')[0]
            z=df2["Products - Southlake"][i].split(':')[1]
            df3=df3.append({'Product-Category':x, 'Product-Sub-Category':y, 'Description':z,'QTY':df2['Qty'][i],'Amount':df2['Amount'][i],'% Cat Sales':df2['% Cat\nSales'][i],'% Total Sales':df2['% Total\nSales'][i]},ignore_index=True)
    
#df3 = df3.astype({'QTY':'int','Amount':'float','% Cat Sales':'float','% Total Sales':'float'})
```

```python
df1.to_csv("Transformed.csv")
```
